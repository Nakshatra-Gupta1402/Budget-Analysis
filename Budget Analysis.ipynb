```python
# Webscrapping the Budget Report 2024
import requests
from bs4 import BeautifulSoup
import bleach


# Fetching text from the webpage
url = 'https://www.livemint.com/economy/budget-2024-25-key-highlights-live-updates-interim-budget-agriculture-infra-fiscal-deficit-nirmala-sitharaman-11706695416199.html'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
s = soup.find("div", class_ = "contentSec" )

content = s.find_all("p")
content = str(content)
clean = bleach.clean(content, tags = [], strip = True)
print(clean)
```

## Text Summarization

```python
from transformers import pipeline
summarizer = pipeline("summarization", model="Azma-AI/bart-large-text-summarizer")
summarizer(clean,truncation=True,max_length=200,min_length=130)[0].get('summary_text')
```

## Sentiment Analysis

```python
## using regex to clean the text
import re
text = re.sub(r'\s+', ' ', clean)
# Remove trailing periods
text = re.sub(r'\.+\s*$', '', text)
# Remove trailing . . .
text = re.sub(r'\. \. \. \.$', '', text)
text
```

```python
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
text =  ' '.join(words)
```

```python
text = re.sub(r"[^.\w\s+]",'',text)
text
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
pipe = pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
a =1
for i in text.split("."):
    print(a, i)
    print(pipe(i))
    print()
    a=a+1
```
